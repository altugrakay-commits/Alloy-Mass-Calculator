```python
import pandas as pd
#Imports the .json file needed to make the calculation using percentages

total_batch_mass = float(input("Please enter the total batch mass in grams (numerical value only): "))
alloy_grade = str(input("Please enter the steel alloy grade (numerical value only): "))
# Requests the user to enter the total batch mass and steel grade, respectively. Line 4 captures user input.

df_data = pd.read_json("steel_data.json")
#Loads the file into the computer´s memory and saves it in the storage container ´df_data.´

target_grade = f"AISI_{alloy_grade}".upper()
#Creates the full grade string by adding the "AISI_" prefix to the user´s input.
```

```python
selected_row_Cr = df_data[df_data.index == 'Cr'][target_grade].item()
try:
	mass_element_Cr = (selected_row_Cr / 100) * total_batch_mass
	print(f"Cr: {mass_element_Cr} grams")
except:
	min_mass_cr = (selected_row_Cr['min'] / 100) * total_batch_mass
	max_mass_cr = (selected_row_Cr['max'] / 100) * total_batch_mass
	print(f"Cr range: {min_mass_cr}-{max_mass_cr} grams")

selected_row_Ni = df_data[df_data.index == 'Ni'][target_grade].item()
try:
	mass_element_Ni = (selected_row_Ni / 100) * total_batch_mass
	print(f"Ni: {mass_element_Ni} grams")
except:
    if (target_grade == 'AISI_430'):
    	print(f"{target_grade} has less than {selected_row_Ni} grams of Ni")
    elif (target_grade == 'AISI_321'):
        max_mass_ni = (selected_row_Ni['max'] / 100) * total_batch_mass
        print(f"{target_grade} has maximum {max_mass_ni} grams only")
    else:
    		min_mass_ni = (selected_row_Ni['min'] / 100) * total_batch_mass
    		max_mass_ni = (selected_row_Ni['max'] / 100) * total_batch_mass
    		print(f"Ni range: {min_mass_ni}-{max_mass_ni} grams")
        #Accounts for missing minimum and maximum numbers, or numbers with strings.
```
